<div style=" padding: 20px"> <h1 style=" margin-bottom:20px">Шаг 3.</h1>
Выбираю топ-10 городов по количеству рейсов   

```python
n = 10
best10 = data_3.sort_values(by = 'average_flights', ascending=False)[:n]
print(best10)
```

<div style=" padding: 20px"> <h1 style=" margin-bottom:20px">Шаг 4. 
Построение графиков

```python
import matplotlib.pyplot as plt
best10.pivot_table(index='city', values='average_flights' ).sort_values(by = 'average_flights', ascending=False).plot( grid=True, figsize=(20, 5),title='топ-10 городов и количество рейсов')
plt.show()
```
