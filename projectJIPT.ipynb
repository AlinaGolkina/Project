```python
query = ''' SELECT *
            FROM log_raw
        '''
log_raw = pd.io.sql.read_sql(query, con = engine)
log_raw
```

```python
log_raw_new = log_raw[['item_topic','event','item_id', 'user_id']].drop_duplicates()
#convert['convert'] = convert['event'].shift(-2)/convert['event'].shift(-1)
convert = log_raw_new.groupby(['item_topic', 'event'])['user_id'].nunique().reset_index()
convert.head(5)
```

```python
view_data = convert[convert['event'] == 'view']
view_data.columns = ['item_topic','event','user_view']
view_data.head(5)
```

```python
click_data = convert[convert['event'] == 'click']
click_data.columns = ['item_topic','event','user_click']
click_data.head(5)
```

```python
data = view_data.merge(click_data, how = 'left', left_on='item_topic', right_on='item_topic')
data.head(5)
```

```python
data['convert'] = data['user_view']/data['user_click']
data = data[['item_topic','user_view', 'user_click','convert']]
data.sort_values('convert', ascending= False)          
```
